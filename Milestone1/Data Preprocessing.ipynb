```python
import pandas as pd
import numpy as np

#Load Dataset
df = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv")
print(df)

#Display number of rows and columns
df.shape

#Display first 5 rows of datset
df.head()

#Check unique values in month column
df['month'].unique()

#Display summary of Dataframe,including data types and non-null values
df.info()

#Check for any row that contains atleast one missing (Nan) value
df[df.isnull().any(axis=1)]

#Region encoding 
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1

#Creating copy of main Dataframe for further processing
dataset=df
dataset.info()
dataset.head

#Converting the 'Region' Column to integer type
dataset[['Region']]=df[['Region']].astype(int)

#Display the basic information after the change 
dataset.head()
dataset.tail()

#Check for missing values after the initial change
dataset.isnull().sum()

#Drop rows with any missing values and reset index to ensure sequential indices and check the first five rows
dataset=dataset.dropna().reset_index(drop=True)
dataset.head()

#Check the count of missing values again (should be 0)
dataset.isnull().sum()

#Drop row at index 122 which was used for region splitting and check the row number 122 after the change
dataset.iloc[[122]]
dataset=dataset.drop(122).reset_index(drop=True)
dataset.iloc[[122]]

#Check column names and remove the trailing spaces 
dataset.columns
dataset.columns=dataset.columns.str.strip()
dataset.columns

#Convert the categorical data into numerical data type
dataset.info()
dataset[['month','day','year','Temperature','RH','Ws']]=dataset[['month','day','year','Temperature','RH','Ws']].astype(int)
dataset.info()
objects=[features for features in dataset.columns if dataset[features].dtypes=='O']
objects
for i in objects:
    if i!='FWI':
        dataset[i]=dataset[i].astype(float)

#Check the Dataframe after the changes done
dataset.info() 
dataset.describe()   
dataset.head() 

#Convert the dataset into csv and excel files
dataset.to_csv('Algerian_forest_fires_dataset.csv',index=False) 
dataset.to_excel('forest_fires_dataset.xlsx',index=False)
data = pd.read_csv("Algerian_forest_fires_dataset.csv")
print(data)


```
