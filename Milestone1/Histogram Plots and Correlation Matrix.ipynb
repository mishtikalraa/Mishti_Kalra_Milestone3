```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("Algerian_forest_fires_dataset_CLEANED.csv")

# Assuming your DataFrame is named 'dataset'
# Histogram of FWI to analyze distribution
plt.figure(figsize=(7,5))
plt.hist(data['FWI'], bins=30)
plt.title('Histogram of FWI')
plt.xlabel('FWI Values')
plt.ylabel('Frequency')
plt.show()

feature_names = data.columns.tolist()
feature_names
for col in feature_names:
    if col != 'FWI':
        plt.figure(figsize=(7,5))

        plt.hist(data[col], bins=30, alpha=0.5, label=col)
        plt.title(f'Histogram Comparison: {col} vs FWI')
        plt.xlabel('Value')
        plt.ylabel('Frequency')
        plt.legend()
        plt.show()

corr_matrix=data.corr(numeric_only=True)
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
